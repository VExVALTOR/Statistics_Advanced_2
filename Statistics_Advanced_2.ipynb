# Question 1: What is hypothesis testing in statistics?  
# Answer:  
Hypothesis testing is a statistical method used to make decisions or inferences about a population based on sample data. It involves formulating two competing hypotheses (null and alternative) and using sample evidence to determine which hypothesis is more likely to be true. It helps in validating assumptions and testing theories scientifically.

# Question 2: What is the null hypothesis, and how does it differ from the alternative hypothesis?  
# Answer:  
The null hypothesis (H₀) is a statement that assumes no effect, no difference, or no relationship exists in the population. It represents the default or status quo assumption.  
The alternative hypothesis (H₁ or Ha) is a statement that contradicts the null hypothesis and proposes that an effect, difference, or relationship does exist.  
For example,  
- H₀: The average height of students = 160 cm  
- H₁: The average height of students ≠ 160 cm  

# Question 3: Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.  
# Answer:  
The significance level (α) is the threshold probability used to decide whether to reject the null hypothesis. Common values are 0.05 or 0.01. It represents the probability of making a Type I error (rejecting a true null hypothesis).  
If the p-value obtained from the test is less than α, we reject H₀; otherwise, we fail to reject H₀. Thus, α controls the strictness of the test.

# Question 4: What are Type I and Type II errors? Give examples of each.  
# Answer:  
- Type I Error: Rejecting the null hypothesis when it is actually true (false positive).  
  Example: Concluding that a medicine works when in reality it does not.  
- Type II Error: Failing to reject the null hypothesis when it is false (false negative).  
  Example: Concluding that a medicine does not work when in reality it does.  

# Question 5: What is the difference between a Z-test and a T-test? Explain when to use each.  
# Answer:  
- Z-test: Used when the population variance is known, and the sample size is large (n > 30). It assumes data follows a normal distribution.  
- T-test: Used when the population variance is unknown, especially for small samples (n ≤ 30). It uses the sample standard deviation instead of population standard deviation.  


```python
#6. Write a Python program to generate a binomial distribution with n=10 and p=0.5, then plot its histogram.

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import binom

n = 10
p = 0.5
sample_size = 10000

binomial_data = np.random.binomial(n, p, sample_size)

plt.figure(figsize=(10, 6))
plt.hist(binomial_data, bins=range(n+2), density=True, alpha=0.7,
         color='skyblue', edgecolor='black', label='Empirical Distribution')

x = np.arange(0, n+1)
theoretical_probs = binom.pmf(x, n, p)
plt.plot(x, theoretical_probs, 'ro-', linewidth=2, markersize=8,
         label='Theoretical Distribution')

plt.title(f'Binomial Distribution (n={n}, p={p})', fontsize=16, fontweight='bold')
plt.xlabel('Number of Successes', fontsize=12)
plt.ylabel('Probability Density', fontsize=12)
plt.xticks(range(n+1))
plt.grid(alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()

print(f"Sample Statistics:")
print(f"Mean: {np.mean(binomial_data):.4f} (Theoretical: {n*p})")
print(f"Variance: {np.var(binomial_data):.4f} (Theoretical: {n*p*(1-p):.4f})")
print(f"Standard Deviation: {np.std(binomial_data):.4f}")
```

```python
'''
#7. Implement hypothesis testing using Z-statistics for a sample dataset in Python. Show the Python code and interpret the results.
Sample data:
49.1, 50.2, 51.0, 48.7, 50.5, 48.8, 50.3, 50.7, 50.2, 49.6,
50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
50.1, 50.3, 50.4, 50.5, 50.7, 49.3, 49.8, 50.2, 50.9, 50.0,
50.3, 50.4, 50.0, 49.7, 50.5, 49.9
'''

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

data = np.array([
    49.1, 50.2, 51.0, 48.7, 50.5, 48.8, 50.3, 50.7, 50.2, 49.6,
    50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
    50.1, 50.3, 50.4, 50.5, 50.7, 49.3, 49.8, 50.2, 50.9, 50.0,
    50.3, 50.4, 50.0, 49.7, 50.5, 49.9
])
population_mean = 50.0
alpha = 0.05

sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
sample_size = len(data)
print(f"Sample size: {sample_size}")
print(f"Sample mean: {sample_mean:.4f}")
print(f"Sample standard deviation: {sample_std:.4f}")

z_statistic = (sample_mean - population_mean) / (sample_std / np.sqrt(sample_size))
print(f"Z-statistic: {z_statistic:.4f}")

p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))
print(f"P-value: {p_value:.6f}")

z_critical = stats.norm.ppf(1 - alpha/2)
print(f"Critical Z-value (α={alpha}): ±{z_critical:.4f}")

print("\n" + "="*50)
print("HYPOTHESIS TEST RESULTS:")
print("="*50)
print(f"Null Hypothesis (H₀): μ = {population_mean}")
print(f"Alternative Hypothesis (H₁): μ ≠ {population_mean}")
print(f"Significance level: α = {alpha}")

if p_value < alpha:
    print("Decision: REJECT the null hypothesis")
    print(f"Reason: p-value ({p_value:.6f}) < α ({alpha})")
else:
    print("Decision: FAIL TO REJECT the null hypothesis")
    print(f"Reason: p-value ({p_value:.6f}) ≥ α ({alpha})")

print(f"Z-statistic: {z_statistic:.4f}")
print(f"Critical region: |Z| > {z_critical:.4f}")

margin_of_error = z_critical * (sample_std / np.sqrt(sample_size))
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error
print(f"95% Confidence Interval: ({ci_lower:.4f}, {ci_upper:.4f})")

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.hist(data, bins=10, alpha=0.7, color='skyblue', edgecolor='black')
plt.axvline(population_mean, color='red', linestyle='--', linewidth=2, label=f'μ₀ = {population_mean}')
plt.axvline(sample_mean, color='green', linestyle='--', linewidth=2, label=f'Sample mean = {sample_mean:.3f}')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Sample Data Distribution')
plt.legend()
plt.grid(alpha=0.3)

plt.subplot(1, 2, 2)
x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x)
plt.plot(x, y, 'b-', label='Standard Normal Distribution')
plt.axvline(z_statistic, color='red', linestyle='-', linewidth=2, label=f'Z-statistic = {z_statistic:.3f}')
plt.axvline(z_critical, color='orange', linestyle='--', linewidth=2, label=f'Critical Z = {z_critical:.3f}')
plt.axvline(-z_critical, color='orange', linestyle='--', linewidth=2, label=f'Critical Z = -{z_critical:.3f}')

x_fill_left = np.linspace(-4, -z_critical, 100)
x_fill_right = np.linspace(z_critical, 4, 100)
plt.fill_between(x_fill_left, stats.norm.pdf(x_fill_left), color='red', alpha=0.3, label='Rejection Region')
plt.fill_between(x_fill_right, stats.norm.pdf(x_fill_right), color='red', alpha=0.3)

plt.xlabel('Z-value')
plt.ylabel('Probability Density')
plt.title('Z-test Visualization')
plt.legend()
plt.grid(alpha=0.3)

plt.tight_layout()
plt.show()

print("\n" + "="*50)
print("INTERPRETATION:")
print("="*50)
print("Based on the Z-test results:")
if p_value < alpha:
    print("• There is sufficient statistical evidence to conclude that")
    print(f"  the population mean is significantly different from {population_mean}")
    print(f"• The sample mean ({sample_mean:.4f}) suggests the true population mean")
    if sample_mean > population_mean:
        print("  is likely GREATER than the hypothesized value")
    else:
        print("  is likely LESS than the hypothesized value")
else:
    print("• There is insufficient evidence to conclude that")
    print(f"  the population mean differs from {population_mean}")
    print("• The observed difference could be due to random sampling variation")

print(f"• The 95% confidence interval suggests the true population mean")
print(f"  is between {ci_lower:.4f} and {ci_upper:.4f}")
```

```python
#8. Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(42)

true_mean = 50
true_std = 5
sample_size = 100

data = np.random.normal(loc=true_mean, scale=true_std, size=sample_size)

sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
standard_error = sample_std / np.sqrt(sample_size)

confidence_level = 0.95
alpha = 1 - confidence_level
degrees_freedom = sample_size - 1

t_critical = stats.t.ppf(1 - alpha/2, degrees_freedom)
margin_of_error = t_critical * standard_error
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

z_critical = stats.norm.ppf(1 - alpha/2)
margin_of_error_z = z_critical * standard_error
ci_lower_z = sample_mean - margin_of_error_z
ci_upper_z = sample_mean + margin_of_error_z

print("=" * 60)
print("NORMAL DISTRIBUTION SIMULATION AND CONFIDENCE INTERVAL")
print("=" * 60)
print(f"True population parameters: μ = {true_mean}, σ = {true_std}")
print(f"Sample size: {sample_size}")
print(f"Sample mean: {sample_mean:.4f}")
print(f"Sample standard deviation: {sample_std:.4f}")
print(f"Standard error: {standard_error:.4f}")
print("\n95% Confidence Intervals:")
print(f"Using t-distribution: ({ci_lower:.4f}, {ci_upper:.4f})")
print(f"Using normal distribution: ({ci_lower_z:.4f}, {ci_upper_z:.4f})")
print(f"Interval width (t-dist): {ci_upper - ci_lower:.4f}")
print(f"True mean captured: {ci_lower <= true_mean <= ci_upper}")

plt.figure(figsize=(15, 10))
plt.subplot(2, 2, 1)
n, bins, patches = plt.hist(data, bins=15, alpha=0.7, color='lightblue', edgecolor='black', density=True)

plt.axvline(true_mean, color='red', linestyle='--', linewidth=2, label=f'True mean (μ={true_mean})')
plt.axvline(sample_mean, color='green', linestyle='--', linewidth=2, label=f'Sample mean (x̄={sample_mean:.2f})')
plt.axvline(ci_lower, color='orange', linestyle=':', linewidth=2, label=f'CI Lower ({ci_lower:.2f})')
plt.axvline(ci_upper, color='orange', linestyle=':', linewidth=2, label=f'CI Upper ({ci_upper:.2f})')

plt.axvspan(ci_lower, ci_upper, alpha=0.2, color='yellow', label='95% Confidence Interval')
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Histogram with 95% Confidence Interval')
plt.legend()
plt.grid(alpha=0.3)

plt.subplot(2, 2, 2)
box = plt.boxplot(data, vert=False, patch_artist=True)
box['boxes'][0].set_facecolor('lightblue')

plt.axvline(ci_lower, color='orange', linestyle=':', linewidth=2)
plt.axvline(ci_upper, color='orange', linestyle=':', linewidth=2)
plt.axvline(sample_mean, color='green', linestyle='--', linewidth=2)
plt.axvline(true_mean, color='red', linestyle='--', linewidth=2)

plt.xlabel('Value')
plt.title('Box Plot with Confidence Interval')
plt.grid(alpha=0.3)

plt.subplot(2, 2, 3)
stats.probplot(data, dist="norm", plot=plt)
plt.title('Q-Q Plot for Normality Check')
plt.grid(alpha=0.3)

plt.subplot(2, 2, 4)
plt.errorbar(x=1, y=sample_mean, yerr=margin_of_error,fmt='o', color='blue', markersize=8, capsize=10, label='Sample mean ± 95% CI')
plt.axhline(y=true_mean, color='red', linestyle='--', label=f'True mean (μ={true_mean})')
plt.axhline(y=sample_mean, color='green', linestyle='-', alpha=0.5)

plt.xlim(0.5, 1.5)
plt.ylim(sample_mean - 3*margin_of_error, sample_mean + 3*margin_of_error)
plt.xticks([])
plt.ylabel('Value')
plt.title('Confidence Interval Visualization')
plt.legend()
plt.grid(alpha=0.3)

plt.tight_layout()
plt.show()

print("\n" + "=" * 60)
print("MULTIPLE SIMULATION ANALYSIS (COVERAGE PROBABILITY)")
print("=" * 60)

n_simulations = 1000
capture_count = 0

for i in range(n_simulations):
    sim_data = np.random.normal(true_mean, true_std, sample_size)
    sim_mean = np.mean(sim_data)
    sim_std = np.std(sim_data, ddof=1)
    sim_se = sim_std / np.sqrt(sample_size)
    sim_t_critical = stats.t.ppf(1 - alpha/2, sample_size - 1)
    sim_ci_lower = sim_mean - sim_t_critical * sim_se
    sim_ci_upper = sim_mean + sim_t_critical * sim_se

    if sim_ci_lower <= true_mean <= sim_ci_upper:
        capture_count += 1

coverage_probability = capture_count / n_simulations
print(f"Number of simulations: {n_simulations}")
print(f"Number of CIs containing true mean: {capture_count}")
print(f"Coverage probability: {coverage_probability:.4f} ({coverage_probability*100:.1f}%)")
print(f"Expected coverage: {confidence_level*100:.1f}%")
print("\n" + "=" * 60)
print("INTERPRETATION")
print("=" * 60)
print("The 95% confidence interval means that if we were to repeat this")
print("sampling process many times, approximately 95% of the calculated")
print("confidence intervals would contain the true population mean.")
print(f"\nIn this case, we are {confidence_level*100:.0f}% confident that")
print(f"the true population mean lies between {ci_lower:.4f} and {ci_upper:.4f}")

if ci_lower <= true_mean <= ci_upper:
    print("✓ The confidence interval successfully captured the true mean!")
else:
    print("✗ The confidence interval did not capture the true mean (this can happen by chance)")

print(f"\nThe multiple simulation analysis shows the actual coverage probability")
print(f"is approximately {coverage_probability*100:.1f}%, which should be close to 95%")
```

```python
#9. Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean.

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

def calculate_and_visualize_z_scores(data, plot_original=True):
    data = np.array(data)
    mean = np.mean(data)
    std = np.std(data, ddof=1)
    z_scores = (data - mean) / std

    if plot_original:
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
    else:
        fig, ax2 = plt.subplots(1, 1, figsize=(10, 6))
        ax1 = None

    if plot_original:
        ax1.hist(data, bins=15, alpha=0.7, color='skyblue', edgecolor='black')
        ax1.axvline(mean, color='red', linestyle='--', linewidth=2,
                   label=f'Mean = {mean:.2f}')
        ax1.axvline(mean + std, color='orange', linestyle=':', linewidth=2,
                   label=f'Mean ± 1σ = {mean + std:.2f}')
        ax1.axvline(mean - std, color='orange', linestyle=':', linewidth=2)
        ax1.set_xlabel('Original Values')
        ax1.set_ylabel('Frequency')
        ax1.set_title('Original Data Distribution')
        ax1.legend()
        ax1.grid(alpha=0.3)

    ax2.hist(z_scores, bins=15, alpha=0.7, color='lightgreen', edgecolor='black')
    ax2.axvline(0, color='red', linestyle='--', linewidth=2, label='Mean = 0')
    ax2.axvline(1, color='orange', linestyle=':', linewidth=2, label='±1σ')
    ax2.axvline(-1, color='orange', linestyle=':', linewidth=2)
    ax2.axvline(2, color='purple', linestyle=':', linewidth=1, alpha=0.7, label='±2σ')
    ax2.axvline(-2, color='purple', linestyle=':', linewidth=1, alpha=0.7)
    ax2.axvline(3, color='brown', linestyle=':', linewidth=1, alpha=0.5, label='±3σ')
    ax2.axvline(-3, color='brown', linestyle=':', linewidth=1, alpha=0.5)

    ax2.set_xlabel('Z-scores')
    ax2.set_ylabel('Frequency')
    ax2.set_title('Standardized Data Distribution (Z-scores)')
    ax2.legend()
    ax2.grid(alpha=0.3)

    plt.tight_layout()
    plt.show()

    print("=" * 60)
    print("Z-SCORE ANALYSIS")
    print("=" * 60)
    print(f"Original data - Mean: {mean:.4f}, Standard Deviation: {std:.4f}")
    print(f"Standardized data - Mean: {np.mean(z_scores):.6f}, Std: {np.std(z_scores, ddof=1):.6f}")
    print(f"Range of Z-scores: [{np.min(z_scores):.4f}, {np.max(z_scores):.4f}]")
    outliers = np.sum(np.abs(z_scores) > 3)
    print(f"Potential outliers (|Z| > 3): {outliers} ({outliers/len(data)*100:.2f}%)")

    return z_scores, {'mean': mean, 'std': std}

if __name__ == "__main__":
    np.random.seed(42)
    sample_data = np.random.normal(loc=50, scale=10, size=1000)
    sample_data = np.concatenate([sample_data, [85, 90, 15, 12]])
    z_scores, stats_info = calculate_and_visualize_z_scores(sample_data)

    print("\n" + "=" * 60)
    print("Z-SCORE INTERPRETATION EXAMPLES")
    print("=" * 60)
    sample_indices = [0, 100, 500, -1, -2]
    for i in sample_indices:
        original_value = sample_data[i]
        z_score = z_scores[i]
        print(f"Value: {original_value:6.2f} → Z-score: {z_score:6.2f} ", end="")
        if abs(z_score) < 1:
            print("(Within 1 standard deviation of mean)")
        elif abs(z_score) < 2:
            print("(Between 1-2 standard deviations from mean)")
        elif abs(z_score) < 3:
            print("(Between 2-3 standard deviations from mean - unusual)")
        else:
            print("(More than 3 standard deviations from mean - potential outlier!)")
    print(f"\nInterpretation in original units (mean = {stats_info['mean']:.2f}, std = {stats_info['std']:.2f}):")
    for z in [-3, -2, -1, 0, 1, 2, 3]:
        original_value = stats_info['mean'] + z * stats_info['std']
        print(f"Z = {z:2.0f} → Original value = {original_value:6.2f}")
```
